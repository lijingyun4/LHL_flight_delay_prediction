```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
from modules.data_gathering import get_working_dataset
```

```python
df = get_working_dataset()
```

```python
df
```

```python
df[['fl_date']] = df[['fl_date']].apply(pd.to_datetime)
df['year'] = df.fl_date.apply(lambda x: x.year)
df['month'] = df.fl_date.apply(lambda x: x.month)
df['day'] = df.fl_date.apply(lambda x: x.day)
```

```python
weather_df = pd.read_csv('/Users/dl/Desktop/lighthouse_midterm/data/WeatherEvents_2016-2021.csv')
```

```python
weather_df.head()
```

```python
weather_df ['origin_city_name'] = weather_df['City'] + ', ' + weather_df['State']
```

```python
weather_df[['StartTime(UTC)','EndTime(UTC)']] = weather_df[['StartTime(UTC)','EndTime(UTC)']].apply(pd.to_datetime)

weather_df ['date'] = weather_df['EndTime(UTC)'].apply(lambda x: datetime.date(x))
```

```python
weather_df.head()
```

```python
weather_df_cleaned = weather_df[['Type', 'origin_city_name','date']]
weather_df_cleaned['fl_date'] = weather_df_cleaned['date'].astype('datetime64')
```

```python
weather_df_cleaned.groupby(['fl_date','origin_city_name'])
weather_df_cleaned = weather_df_cleaned.drop_duplicates(subset = ['fl_date','origin_city_name'])
```

```python
merge = df.merge(weather_df_cleaned,how='left', on=['fl_date','origin_city_name'])
```

```python
merge
```

```python
merge = merge.drop('date', 1)
```

```python
merge2 = merge.dropna(subset=['fl_date'])
```

```python
merge2[['arr_delay','carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']] = merge2[['arr_delay','carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']].fillna(0)
```

```python
merge2
```

```python
merge2.isna().sum()
```

```python
merge2['Type'] = merge2['Type'].fillna('other')
```

```python
merge2 = merge2.drop_duplicates()
```

```python
merge2 = merge2.reset_index(drop = True)
```

```python
merge2 = merge2.dropna(subset=['tail_num'])
```
